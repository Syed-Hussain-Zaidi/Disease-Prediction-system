```python
numeric_columns = ['Age','Sex','Chest pain type','BP','Cholesterol','FBS over 120','EKG results','Max HR','Exercise angina','ST depression','Slope of ST','Number of vessels fluro','Thallium']
for i in numeric_columns:
    sns.histplot(df[i],bins = 30, kde = True)
```
